# Long to wide table format
### pivot
### pivot table

```python
# Long은 그냥 일반 표라고 생각하고 wide는 표를 정리한 것이라고 생각하면 편할듯
air_quality_no2 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_long.csv", index_col = 'date.utc')
# index_col로 파일을 열 때 index 설정해주기 
# a.to_csv(r"C:\Users\polit\python_Pandas\air_quality_no2.csv")
```

```python
no2 = air_quality_no2[air_quality_no2['parameter'] == 'no2']
```

```python
no2_subset = no2.sort_index().groupby(['location']).head(2)
no2_subset
```

```python
# 피봇테이블로 바꿔주기 
no2_subset.pivot(columns = 'location', values = 'value') 
```

```python
no2_subset.pivot(columns = 'location', values = 'value').plot()
```

# Pivot table

```python
air_quality_no2.pivot_table(index = 'location', columns = 'parameter', values = 'value', aggfunc = 'mean', margins = False)
# aggfunc : 셀에 여래개의 값이 있을 떄 연산을 넣음. sum, mean...
# margin : All 합계를 보여줌
```

```python
# pivot table과 기본적으로 같은 연산을 함 보여주는 방식만 다름 
air_quality_no2.groupby(['location', 'parameter'])['value'].mean()
```

# Wide to long format

```python
# 정리한 표를 원래 표로 되돌려주는 느낌
no2_pivot = no2.pivot(columns = 'location', values = 'value').reset_index()
no2_pivot
```

```python
no2_pivot.melt(id_vars = 'date.utc', value_vars = ['BETR801'])
# id_vars : 기준 컬럼 설정. 같이 녹지 않음 
# value_vars : 녹일(사용할) 컬럼만 지정
```
