# JOIN

### 조인하고자 하는 각 테이블의 기준 컬럼 이름이 같을 때

```python
air_quality_stations = pd.read_csv(r"https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_stations.csv")
air_quality_stations.to_csv(r"C:\Users\polit\python_Pandas\air_quailty_stations.csv")
```

```python
air_quality_parameters = pd.read_csv(r"https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_parameters.csv")
air_quality_parameters.to_csv(r"C:\Users\polit\python_Pandas\air_quailty_parameters.csv")
```

```python
stations_coord = air_quality_stations
stations_coord.head()
```

```python
air_quality.head(3)
```

```python
# location컬럼을 기준으로 left join을 할 것이다. / how의 default는 inner
air_quality = pd.merge(air_quality, stations_coord, how = 'left', on = 'location')
air_quality.head(3)
```

### 기준이 되는 컬럼의 이름이 다를 때

```python
pd.merge(air_quality, air_quality_parameters, how = 'left', left_on = 'parameter', right_on = 'id')
```
