## SQL

### SELECT Name
### FROM titanic
### WHERE Age > 35

### .loc[row조건, column명] = location
### .iloc[row[:] or 인덱스, column[:] or 인덱스] = index location

```python
import pandas as pd

# 데이터 가져오기 
titanic = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv")
```

# .loc[]

```python
# 40세 이상인 승객들의 이름만 추출하기
titanic.loc[titanic['Age'] > 40, "Name"]
```

# .iloc[]

```python
# 0 ~ 9번쨰 행과 0~2번쨰 열을 추출하기 
titanic.iloc[0:10, 0:3]
```

# 데이터 수정하기


```python
 # 0~ 4번째 행과 4번쨰 열에 해당하는 값을 NULL로 수정하기(인덱스는 0부터 시작 ==> 3은 4번쨰 열)
titanic.iloc[0:5, 3] = "NULL"

# 출력하기
titanic 
```
