# 차트 그려보기 .plot()

```python
# 라인 차트 크리기(데이터프래임에서 뽑아보기)

air_quality.plot() # 기본이 .line() 차트
```

```python
# 시리즈를 이용해 뽑아보기 
air_quality['station_paris'].plot() 
```

# 다양한 차트 그려보기 

```python
air_quality.plot.scatter(x = 'station_london', y = 'station_paris', alpha = 0.5)
# .alpha 투명도 주기 
```

```python
air_quality.plot.box()
```

```python
air_quality.plot.area(subplots = True)

# subplots = True 각 축에 대해서 따로 그려줌
```
