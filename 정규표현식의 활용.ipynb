# Series.str.contains 

```python
df= pd.DataFrame(['Mouse', 'dog', 'house and parrot', '23', 'og'], columns = ['values'])
df

# df = pd.DataFrame({'values' : ['Mouse', 'dog', 'house and parrot', '23']}) # 같은 의미
```

```python
df['v1'] = df['values'].str.contains('dog')
df
```

```python
# str.contains() : 정규표현식 regex = True가 default

df['v2'] = df['values'].str.contains('og', regex = False) 
df
```

```python
# str.contains() : 대소문자 구분 case = True가 default / False : 대소문자 신경 안씀
df['v3'] = df['values'].str.contains('oG', regex = False, case = True) 
df
```

```python
# regex 확인해보기 
# \D : 숫자가 아닌 모든 글자
# \d : 모든 숫자
df['v4'] = df['values'].str.contains('\D', regex = True)
df['v5'] = df['values'].str.contains('\D', regex = False)
df
```

```python
df2 = pd.DataFrame(['40', '40.0', '41', '41.0', '35'], columns = ['values'])
df2
```

```python
# '.' : 모든 캐릭터
# 0앞에 어떤 캐릭터가 있어도 True
df2['v1'] = df2['values'].str.contains('.0', regex = True) 

df2['v2'] = df2['values'].str.contains('.0', regex = False)
df2
```

# Series.str.extract

```python
df3 = pd.DataFrame(['asda65', 'bqwr2', 'cas3',"aasf 65"], columns = ['values'])
df3
```

```python
# 앞에 a 또는 b가 있을 것이고 뒤에 숫자가 있을 것이다.
# ()()로 묶으면 따로 추출하겠다는 의미
df3['values'].str.extract(r'([ab ])(\d)') # 이거로 눌값 만들자
```

```python
# ? : 앞에 있는 문자가 있어도 되고 없어도 된다. 뒤에만 필수
df3['values'].str.extract(r'([ab])?(\d)') 
```

```python
# ()로 묶지 않으면 추출하지 않음 but 필터링은 함 
# expand : 데이터프래임으로 추출하겠다. True가 default
df3['values'].str.extract(r'[ab](\d)', expand = False) 
```

```python
df3['values'].str.extract(r'([ab]\d)') 
```
