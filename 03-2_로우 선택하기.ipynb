# 연령 35세 초과인 승객 추출하기

```python
import pandas as pd

# 데이터 가져오기 
titanic = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv")
```

```python
above_35 = titanic[titanic['Age'] > 35] 
above_35
```

```python
print(type(titanic[titanic['Age'] > 35]))
```

# Pclass 선택하기

### .isin([ ])
### OR, AND


```python
# < OR >

# .isin() 은 Series가 []리스트 안에 있는 값을 가지고 있다면 True 아니면 False

Pclass_2_or_3 =titanic['Pclass'].isin([2, 3])
Pclass_2_or_3
# (titanic['Pclass'] == 3) || (titanic['Pclass'] == 3) 이런식으로도 표현 가능 
```

```python
print(type(titanic['Pclass'].isin([2, 3])))
```

```python
# Pclass가 2 혹은 3인 데이터 추출 
titanic[titanic['Pclass'].isin([2, 3])]
```

```python
# < AND >

# Pclass가 3이면서 Age가 30인 승객들 추출
titanic[(titanic['Pclass'] == 3) & (titanic['Age'] == 30)]
```
